# Q - 2. - Secant Method

## Setup

```python
def f_line(f, x, x_n, x_nn):
    temp = x_n
    slope = (f(x_n) - f(x_nn))/(x_n - x_nn)
    x_n = x_n - f(x_n) / slope
    x_nn = temp
    return slope * (x - x_n) + f(x_n), x_n, x_nn

def update_plot():
    x_n = 3
    x_nn = 2
    ax.clear()
    ax.plot(x, f(x), label='f(x)', color='g')
    ax.axhline(0, color='gray', lw=0.5)
    colors = ['g', 'r', 'b', 'g']
    for i, color in enumerate(colors):

        if i >= 1:
            ax.plot(x, f_l, '--', lw=1, color=color)
        
        ax.plot(x_n, f(x_n), 'kd', label=r'$f(x_{}={:.2f})={:.3f}$'.format(i, x_n, f(x_n)))
        ax.vlines(x_n, 0, f(x_n), linestyle=':', lw=1)
        f_l, x_n, x_nn = f_line(f, x, x_n, x_nn)

        
    fig = plt.figure()
    ax.set_ylim(-9, 9)
    ax.set_xlim(-4, 4)
    ax.legend(loc=4)
    plt.show()
```

# Q - 3. - Bisection Method

## Setup

```python
def bisect(f, a, b):
    mid = (a+b)/2
    result = f(mid)
    if (result*f(a) < 0):
        b = mid
    else:
        a = mid
    
    return a, b, mid


def update_plot(a=-3, b = 3):
    ax.clear()
    ax.plot(x, f(x), label='f(x)', color='g')
    ax.axhline(0, color='gray', lw=0.5)
    colors = ['g', 'r', 'b', 'g']
    for i, color in enumerate(colors):

        if i >= 1:
            ax.scatter(mid, f(mid), color=color, label=r'$f(x_{}={:.2f})={:.3f}$'.format(i, mid, f(mid)))
            ax.vlines(mid, 0, f(mid), linestyle=':', lw=1)

        a, b, mid = bisect(f, a, b)

        
    fig = plt.figure()
    ax.set_ylim(-9, 9)
    ax.set_xlim(-4, 4)
    ax.legend(loc=4)
    plt.show()
```

## 1.)

```python
def f(x):
    return x**2 - 5
fig, ax = plt.subplots(1, figsize=(13, 6.5))
update_plot()
```

# Q - 7. - Exact roots

## Setup

```python
x = symbols('x')
def get_solutions(f, x):
  solutions = solve(f(x), x, dict=True)
  print('For expression, {}: '.format((f(x))))
  for solution in solutions:
      print(solution[x].evalf())
```

## 1.)

```python
def f(x):
    return x**2 - 5

get_solutions(f, x)
```
